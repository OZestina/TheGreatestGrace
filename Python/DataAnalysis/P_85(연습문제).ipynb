```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'data/drinks.csv'
drinks = pd.read_csv(file_path)
drinks['continent'] = drinks['continent'].fillna('OT')
drinks.info()
```

```python
#1번

drinks.groupby('continent')['wine_servings'].mean()
```

```python
#2번

drinks['total_servings'] = drinks['beer_servings'] + drinks['wine_servings'] + drinks['spirit_servings'] 
drinks.head()
```

```python
#3번

alcohol_mean = drinks['total_litres_of_pure_alcohol'].mean()

total_alcohol = drinks['total_litres_of_pure_alcohol'].sum()
total_country = len(drinks['total_litres_of_pure_alcohol'])

print(alcohol_mean)
# print(total_alcohol)
# print(total_country)
# print(total_alcohol / total_country)

continent_mean = drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean()
#print(continent_mean)

#continent_low = continent_mean[continent_mean < alcohol_mean]
#print(continent_low) #AF, AS, OC

continent_low = continent_mean[continent_mean < alcohol_mean].index.tolist()
print(continent_low)


# a = drinks[['country','spirit_servings', 'continent']]['continent']== ('AF' or 'AS' or 'OC')

country_low = drinks.loc[drinks.continent.isin(continent_low)]
# print(country_low)
# len(country_low) #113

#country_low = drinks.loc[drinks['continent']=='AF' or 'AS' or 'OC']


#country_low['spirit_servings'].idxmax()
#print(country_low.groupby('country').spirit_servings.idxmax())

mostMin = country_low.loc[country_low['spirit_servings'].idxmax()]
# mostMin.head(5)
# mostMin.head(1)

print(mostMin)

print(mostMin['country'])
```

```python
total_mean = drinks.total_litres_of_pure_alcohol.mean()
continent_mean = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()

continent_under_mean = continent_mean[continent_mean <= total_mean].index.tolist()
#print(continent_under_mean)

df_continent_under_mean = drinks.loc[drinks.continent.isin(continent_under_mean)]
#print(df_continent_under_mean)

most_spirit_under_mean = df_continent_under_mean.loc[df_continent_under_mean['spirit_servings'].idxmax()]
print(most_spirit_under_mean )

most_spirit_under_mean['country']

len(df_continent_under_mean)
```
